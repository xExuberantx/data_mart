```python
import pandas as pd
clean_weekly_sales = pd.read_csv(r'C:\Users\Kamil\Desktop\data_mart_pd\clean_weekly_sales.csv')
clean_weekly_sales.head()
```

```python
sales_by_platform = clean_weekly_sales.groupby(['week_number', 'platform'])['sales'].sum()
sales_by_platform = sales_by_platform.reset_index()
print(sales_by_platform)
```

```python
sbp = sales_by_platform.pivot_table(index='week_number', columns='platform', values='sales')
sbp
```
